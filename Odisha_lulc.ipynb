```python
import geemap.geemap as geemap
import ee
```

```python
ee.Authenticate()
ee.Initialize(project='neon-feat-313910')
base = geemap.Map()
```

```python
Odisha = ee.FeatureCollection('projects/neon-feat-313910/assets/Odisha')
```

```python
worldcover = ee.ImageCollection("ESA/WorldCover/v200").mean().clip(Odisha)
base.addLayer(worldcover,{'min':10,'max':90, 'palette':['006400','ffbb22','ffff4c','f096ff','fa0000', 'b4b4b4',  '0064c8',  '0096a0']}, 'ESA WorldCover')
base.add_legend(title = "LULC Legend", legend_dict = {
                                                    'Tree Cover':'006400',
                                                    'Shrub Land':'ffbb22',
                                                    'Grass Land':'ffff4c',
                                                    'Crop Land' :'f096ff',
                                                    'Built Up': 'fa0000',
                                                    'Sparse Vegetation':'b4b4b4',
                                                    'Permanent Water Bodies':'0064c8',
                                                    'Wetlands': '0096a0'
},position = "bottomleft")
```

```python
training_points = worldcover.sample(**{'region':Odisha.geometry(),
                                    'scale':20,
                                    'numPixels':2000,
                                    'seed':1,
                                    'geometries':True})
base.addLayer(training_points, {}, 'Training Points')
base.centerObject(Odisha, 7)
base
```
